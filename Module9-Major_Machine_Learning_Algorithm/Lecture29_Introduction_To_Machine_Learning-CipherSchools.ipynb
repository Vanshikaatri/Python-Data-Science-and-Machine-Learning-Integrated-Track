# What is ML?

**Definition**-
Machine learning is subset of AI that uses statistical methods to allow machines to improve their results with the amount of training data fed to it.

Screenshot 2024-06-27 150412.png

**Supervised Learning**:

Supervised learning is when we teach or train the machine using data that is well-labelled. Which means some data is already tagged with the correct answer.

**Unsupervised Learning**:

Unsupervised learning is a branch of machine learning that deals with unlabeled data.

# What is Data Science?

**Definition**:

* Data Science is a field of study invloving statistical tools and techniques to extract meaningful insights from data.
* The data from which these insights are extracted can come from various sources including databases, business transactions, sensors and more.
* Data science used complex machine learning algorithms to build predictive models.
