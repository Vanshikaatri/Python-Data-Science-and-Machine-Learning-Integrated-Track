# Introduction to Matplotlib and Seaborn

**Definition** :


*   **Matplotlib** - Matplotlib is a comprehensive library for static, animated, and creative visualizations in Python. It provides an object-oriented API for embedding plots into applications general-purpose GUI toolkits.
*   **Seaborn** - Seaborn is a Python data visualization library based on Matplotlib. It provides a high level interface for drawing attractive and informative statistical graphics

**Use Case in Real Life**:


*   **Data Exploration**- Create various plots to visulaize data distributions and relationships between variables during the exploratory data analysis (EDA) phase.
*   **Statistical Analysis**- Use visualizations to understand statistical properties of datasets, such as distribution plots, histograms and pair plots to identify correlation and patterns.


*   **Business Reporting**- Generate business reports with visualizations that provide insights into sales performance, customer behaviour

```python
import matplotlib.pyplot as plt        # pyplot is sub module

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Creating Line Plot
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Adding style to the Line
plt.plot(x, y, linestyle = "dashed")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
# Shortcut for styling
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Adding style to the Line, line color, line width, markers with shortcuts
plt.plot(x, y, ls= "-", color = "r", lw = "2", marker= "*", mfc = "b", ms = '20')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

# Creating a Simple Scatter Plot with Matplotlib

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Adding style to the Line
plt.scatter(x, y, color = "y")
plt.xlabel('sales')
plt.ylabel('Price')
plt.title('Simple Scatter Plot')
plt.show()
```

# Creating a Simple Bar Plot with Matplotlib

```python
# Data
categories = ['A', 'B', 'C', 'D']
values = [4,7,1,8]

# Creating a bar plot
plt.barh(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Plot')
plt.show()
```

```python
# Data
products = ['PC', 'TV', 'Ref', 'Micro']
sales = [4, 7, 1, 8]

# Creating a bar plot
plt.bar(products, sales, color = 'black', width = 0.1)         # while using barh we will replace width with height
plt.xlabel('products')
plt.ylabel('sales(in 1000s)')
plt.title('Simple bar plot')
plt.show()
```

# Creating a Simple Histogram with Matplotlib

```python
# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

# Creating a histogram
plt.hist(data, bins=4)       # bins = binsize means that we are taking 4 intervals in a graph
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Simple Histogram')
plt.show()
```

# Creating SubPlots with Matplotlib

```python
# Data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

# Creating subplots
fig, axs = plt.subplots(2)

# Plotting
axs[0].plot(x, y1)
axs[0].set_title('First Plot')
axs[1].plot(x, y2, 'tab:orange')
axs[1].set_title('Second Plot')

# Displaying the plots
plt.tight_layout()
plt.show()
```

# Adding annotations with Matplotlib

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Creating a plot with annotations
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot with Annotations')
plt.annotate('Peak', xy=(5,11), xytext=(4,10),
             arrowprops=dict(facecolor="black", shrink=0.05))
plt.show()
```

# Creating a Simple Line plot with Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]

# Creating a line plot
sns.lineplot(x=x1, y=y1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot with Seaborn')
plt.show()
```

# Creating a Simple Scatter Plot with Seaborn

# Creating a Pair Plot with Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Loading the Iris dataset
iris = load_iris()
iris_data = sns.load_dataset("iris")

# Creating a pair plot
sns.pairplot(iris_data, hue="species")
plt.title('Pair plot with Seaborn')
plt.show()
```
